# Explorar dados Gold via `data_reader`

Consultas ao SQLite (`data/app.db`) usando [`rf_lake.data_reader.Database`](../rf_lake/data_reader/database.py).

**Pré-requisitos:** `pip install -e .` no venv e lake populado (`python run_lake.py daily` ou `backfill`).

Ajuste `START` e `END` na célula seguinte.

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if (ROOT / "rf_lake").is_dir():
    PROJECT_ROOT = ROOT
elif (ROOT.parent / "rf_lake").is_dir():
    PROJECT_ROOT = ROOT.parent
else:
    raise RuntimeError("Abra o notebook na raiz do projeto ou na pasta notebooks/")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from rf_lake.data_reader import Database
from rf_lake.settings import DB_PATH, DATA_START_DATE

db = Database()
print("PROJECT_ROOT:", PROJECT_ROOT)
print("DB:", DB_PATH)
print("DATA_START_DATE:", DATA_START_DATE)
```

```python
START = "2026-01-02"
END = "2026-01-15"


def preview(df, title: str, n: int = 5):
    print("\n" + "=" * 60)
    print(f"{title} | shape: {df.shape}")
    print("=" * 60)
    display(df.head(n))
    print("\ndtypes:")
    print(df.dtypes)
```

## Mercado secundário e liquidações

```python
preview(db.get_mercado_secundario(start_date="2026-05-13", end_date="2026-05-13"), "Mercado secundário")
preview(db.get_liquidacoes_mercado(start_date=START, end_date=END), "Liquidações mercado")
preview(
    db.get_mercado_secundario_com_liquidacoes(start_date=START, end_date=END),
    "Mercado + liquidações",
)
```

## BMF, leilões e cadastros

```python
preview(db.get_ajustes_bmf(start_date=START, end_date=END), "Ajustes BMF")
preview(db.get_leiloes(start_date="2026-01-01", end_date="2026-05-15"), "Leilões")
preview(db.get_titulos_publicos(), "Títulos públicos (cadastro)")
preview(db.get_contratos_bmf(), "Contratos BMF")
```

## IPCA, projeções e feriados

```python
preview(
    db.get_ipca_indice(start_month=f"{START[:7]}-01", end_month=f"{END[:7]}-01"),
    "IPCA índice",
)
preview(db.get_ipca_indice_last_months(12), "IPCA últimos 12 meses")
preview(
    db.get_projecoes(last=2),
    "Projeções",
)
preview(db.get_feriados(), "Feriados")
```

## Metadados e uma data exata

```python
preview(db.get_schema_migrations(), "Schema migrations")
preview(db.get_job_runs(start_date=START, end_date=END), "Job runs")

REF = START
preview(db.get_mercado_secundario(ref_date=REF), f"Mercado secundário @ {REF}")
```
